```python
import re
from urllib.request import urlopen

page = urlopen("http://www.pythonchallenge.com/pc/def/equality.html")
body = page.read().decode('ascii')
text = body.split("<!--\n")[-1]
text = text.split("-->")[0]

# Match a lower case letter, surrounded by exactly 3 upper
# case characters on each side.
pattern = "[^A-Z][A-Z]{3}[a-z][A-Z]{3}[^A-Z]"
print("".join(match[4] for match in re.findall(pattern, text)))
```

# Level 6

```python
import re
import sys
import tempfile
import zipfile
from io import BytesIO
from urllib.request import urlopen

# Download the zip file into an in-memory file (BytesIO)
# (you can't pass what urlopen returns to ZipFile because it isn't seekable)
zip_content = urlopen("http://www.pythonchallenge.com/pc/def/channel.zip").read()
z = zipfile.ZipFile(BytesIO(zip_content))

nothing = 90052
while True:
    filename = str(nothing)+".txt"
    
    # Extract the file's comment and output
    comment = z.getinfo(filename).comment
    sys.stdout.write(comment)
    
    # Find the next file to open.
    body = z.open(filename).read().decode('ascii')
    m = re.match("Next nothing is (\d+)", body)
    if m:
        nothing = int(m.group(1))
    else:
        break
```

# Level 7

"in the air" -> "oxygen"

```python
from urllib.request import urlopen
from PIL import Image  # for Python 3, "pip install pillow" to get PIL

im = Image.open(urlopen("http://www.pythonchallenge.com/pc/def/oxygen.png"))

# The bar of grey squares is about half way down the image, so extract a 
# single row half way down. Also, the row of squares doesn't go the whole way, 
# so cut a bit off the end.
x_max, y_max = im.size
y_half = y_max // 2
cropped_im = im.crop((0, y_half, x_max-20, y_half + 1))

# Now move through the pixels for the row. 
# The squares change every 7 pixels, only sample the row every 7 pixels.
# Each pixel is represented as (r, g, b, a). Red, green and blue are the 
# same for grey so just read out the red value.
# Then convert it the red value to an ASCII character.
row = cropped_im.getdata()
print("".join(chr(row[i][0]) for i in range(0, len(row), 7)))
```
